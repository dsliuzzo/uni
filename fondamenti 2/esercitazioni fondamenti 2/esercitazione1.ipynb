```python
"""
si scriva una funzione "conta_vocali" che riceve in ingresso una stringa 
e ritorna il numero di vocali in essa presenti (sia maiuscole che minuscole)
"""

def conta_vocali(s):
    vocali = "aeiou"
    cont = 0
    s2 = s.lower()
    for c in s2:
        if c in vocali:
            cont += 1
    return cont


stringa = "aswiuiewqalqj"
print(conta_vocali(stringa))


# versione più compatta
def conta_vocali_v2(s):
    vocali = "aeiou"
    return len([c for c in s.lower() if c in vocali])

# altra versione più compatta
def conta_vocali_v3(s):
    vocali = 'aeiou'
    return sum(1 for c in s.lower() if c in vocali)
```

```python
'''
si scriva una funzione "nessuna_vocale_pari" che riceve in ingresso una stringa
e ritorna True se la stringa non contiene vocali in posizione pari, False altrimenti
'''

def nessuna_vocale_pari(s):
    vocali = "aeiou"
    for i in range(0, len(s), 2):
        if s[i].lower() in vocali:
            return False
    return True


def nessuna_vocale_pari_v2(s):
    vocali = "aeiou"
    return all(s[i].lower() not in vocali for i in range(0, len(s), 2))

def nessuna_vocale_pari_v3(s):
    vocali = "aeiou"
    return all(c.lower() not in vocali for c in s[::2])
```

```python
"""
si scriva una funzione "parola_piu_lunga" che riceve in ingresso una stringa (frase)
e restituisce la parola più lunga presente nella frase
"""

def parola_piu_lunga(s):
    parole = s.split() #costruiamo la lista parole
    max_tmp = 0
    parola_max = ""
    for p in parole:
        if len(p) > max_tmp:
            #allora aggiorniamo
            max_tmp = len(p)
            parola_max = p
    return parola_max

def parola_piu_lunga_v2(s):
    parole = s.split()
    return max(parole, key=len)


stringa = "ciao come stai tutto bene"
print(parola_piu_lunga(stringa))
print(parola_piu_lunga_v2(stringa))
```

```python
"""
si scriva una funzione "rollup" che riceve in input una lista l di lunghezza n
(con n pari) e restituisce una nuova lista di lunghezza n/2 il cui i-esimo elemento
è pari a l[2*i]+l[2*i+1]
"""

def rollup(l):
    ret = []
    for i in range(len(l)//2):
        ret.append(l[2*i] + l[2*i+1])
    return ret

def rollup_v2(l):
    ret = [0]*(len(l)//2)
    for i in range(len(ret)):
        ret[i] = l[2*i] + l[2*i+1]
    return ret
```

```python
"""
si scriva una funzione "alternati" che riceve in input una lista
e restituisce True se e solo se la lista contiene valori alternati pari e dispari,
cioè non ci sono due valori pari o due valori dispari consecutivi
"""

def alternati(l):
    prec_pari = l[0] % 2 == 0
    for i in range(1, len(l)):
        if l[i]%2 == 0 and prec_pari or l[i]%2!=0 and not prec_pari:
            return False
        prec_pari = not prec_pari
    return True
```

```python
"""
si scriva una funzione "costruisci_lista" che riceve in input una lista di interi L1
e restituisce una lista L2 di pari lunghezza, il cui generico i-esimo elemento vale:
- se la media degli elementi di L1 con indice maggiore o uguale a i è maggiore o
uguale a L1[i], allora L2[i] è pari a tale media
- altrimenti L2[i] è uguale alla differenza tra la somma degli elementi di sinistra
di L1[i] e la somma degli elementi a destra di L1[i]
"""


def costruisci_lista(L1):
    L2 = []
    for i in range(len(L1)):
        somma = sum(L1[i:])
        num_elementi = len(L1)-i+1
        media = somma/num_elementi
        if media >= L1[i]:
            L2.append(media)
        else:
            somma_sinistra = sum(L1[:i])
            somma_destra = somma - L1[i] #sum(L1[i+1:])
            L2.append(somma_sinistra - somma_destra)
    return L2
        
```
