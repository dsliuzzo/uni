```python
def verifica(self):
    if self._lunghezza < 3:
        return False
    return self._verifica_ric(self._testa, self._testa.successivo, self._testa.successivo.successivo)


def _verifica_ric(self, prec, corr, succ):
    if succ is None:
        return False
    # siamo su un nodo interno
    if corr.info == max(prec.info, succ.info):
        return True
    return self._verifica_ric(corr, succ, succ.successivo)
```

```python
#3.1

def verifica(self):
    for p in self._listaPersonale:
        tipo = p.get_tipo()
        for nome_sotto in p.get_unita_sotto_controllo():
            altro_p = _trova_personale(nome_sotto)
            altro_tipo = altro_p.get_tipo()
            if altro_tipo != tipo:
                return False
    return True


def _trova_personale(self, nome):
    for p in self._listaPersonale:
        if p.get_nome() == nome:
            return p
    return None
```

```python
#3.2

def medico_frequente_con_ruolo(self, ruolo):
    num_interventi_max = 0
    nome_medico_max = ""
    for p in self._listaPersonale:
        if p.get_tipo():
            cont_interventi = 0
            for i in self._listaInterventi:
                ite = Iteratore(i.get_equipe())
                itr = Iteratore(i.get_ruoli())
                while not ite.finito():
                    nome_corr = ite.next()
                    ruolo_corr = itr.next()
                    if nome_corr == p.get_nome() and ruolo_corr == ruolo:
                        cont_interventi += 1
            if cont_interventi > num_interventi_max:
                num_interventi_max = cont_interventi
                nome_medico_max = p.get_nome()
    return nome_medico_max 
    
```
