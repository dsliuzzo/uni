```python
""" si scriva una funzione verifica_matrice che riceve in input
una matrice di interi e restituisce True sse le colonne pari
sono ordinate in modo non crescente e le colonne dispari sono
ordinate in modo non decrescente. """

def verifica_matrice(M):
    for j in range(len(M)):
        if j % 2 == 0: #colonna di indice pari
            for i in range(len(M)-1):
                if M[i][j] < M[i+1][j]:
                    return False
        else: # colonna di indice dispari
            for i in range(len(M)-1):
                if M[i][j] > M[i+1][j]:
                    return False
    return True


def verifica_matrice_v2(M):
    for j in range(len(M[0])):
        if j % 2 == 0: #colonna di indice pari
            if not all(M[i][j] >= M[i+1][j] for i in range(len(M)-1)):
                return False
        else: # colonna di indice dispari
            if not all(M[i][j] <= M[i+1][j] for i in range(len(M)-1)):
                return False
    return True

def verifica_matrice_v3(M):
    for j in range(len(M[0])):
        if j % 2 == 0: #colonna di indice pari
            if any(M[i][j] < M[i+1][j] for i in range(len(M)-1)):
                return False
        else: # colonna di indice dispari
            if any(M[i][j] > M[i+1][j] for i in range(len(M)-1)):
                return False
    return True


```

```python
""" si scriva una funzione filtra_matrice che riceve in input 
una matrice di interi e restituisce una lista contenente tutti
gli elementi presenti sulle righe di inidice pari che 
contengono solo elementi dispari"""

def filtra_matrice(M):
    ret = []
    for i in range(0, len(M), 2):
        if tutti_dispari(M[i]):
            ret.extend(M[i]) #per aggiungere tutti gli elem senza creare più livelli
    return ret

def tutti_dispari(L):
    for elem in L:
        if elem % 2 == 0:
            return False
    return True
            
```

```python
""" si scriva una funzione a_strisce che riceve in input una matrice 
di interi e restituisce True sse la matrice è a strisce orizzontali,
cioè se le righe di indice dispari contengono solo zeri, mentre
le righe pari non contengono zeri"""

def a_strisce(M):
    for i in range(0, len(M), 2):
        if 0 in M[i]:
            return False
    for i in range(1, len(M), 2):
        for elem in M[i]:
            if elem != 0:
                return False
    return True

def a_strisce_v2(M):
    for i in range(len(M)):
        for j in range(len(M[0])):
            if i%2==0 and M[i][j]==0 or i%2==1 and M[i][j]!=0:
                return False
    return True
```

```python
""" scrivere una funzione ricorsiva che data una stringa s
restituisca la stringa ottenuta da s eliminando le vocali
es: se s = 'pippo', ret = 'ppp'"""

def elimina_vocali(s):
    if len(s) == 0: return s #condizione uscita generale
    return elimina_vocali_ric(s) #init metodo ricorsivo

def elimina_vocali_ric(s_corr):
    vocali = "aeiouAEIOU"
    if len(s_corr) == 0: return s_corr #condizione uscita specifica
    if s_corr[0] in vocali:
        return elimina_vocali_ric(s_corr[1:])
    return s_corr[0] + elimina_vocali_ric(s_corr[1:])


def elimina_vocali_v2(s):
    if len(s) == 0: return s #condizione uscita generale
    return elimina_vocali_ric(s, "") #init metodo ricorsivo

def elimina_vocali_ric(s_corr, s_ret):
    vocali = "aeiouAEIOU"
    if len(s_corr) == 0: return s_ret
    if s_corr[0] not in vocali:
        s_ret += s_corr[0]
    return elimina_vocali_ric(s_corr[1:], s_ret)

```

```python
""" scrivere una funzione ricorsiva che data una lista di 
interi l restituisce la somma dei soli numeri pari"""

def somma_pari(l):
    if len(l) == 0: return 0
    return somma_pari_ric(l, 0) #lista corrente, somma_pari

def somma_pari_ric(l_corr, somma_pari_corr):
    if len(l_corr) == 0: return somma_pari_corr
    if l_corr[0] % 2 == 0:
        somma_pari_corr += l_corr[0]
    return somma_pari_ric(l_corr[1:], somma_pari_corr)


def somma_pari_v2(l):
    if len(l) == 0: return 0
    return somma_pari_ric(l, 0, 0) #lista, indice corr, somma_pari corr

def somma_pari_ric(l, indice_corr, somma_pari_corr):
    if indice_corr == len(l): return somma_pari_corr
    if l[indice_corr] % 2 == 0:
        somma_pari_corr += l[indice_corr]
    return somma_pari_ric(l, indice_corr+1, somma_pari_corr)


def somma_pari_v3(l):
    if len(l) == 0: return 0
    return somma_pari_ric(l)


def somma_pari_ric(l):
    if len(l) == 0: return 0 #if l == []: return 0
    if l[0] % 2 == 0:
        return l[0] + somma_pari_ric(l[1:])
    return somma_pari_ric(l[1:])
    
```

```python
""" scrivere una funziona ricorsiva che, data una stringa s e
un carattere c, ritorna il numero di occorrenze di c in s"""

def conta_occorrenze(c, s):
    if len(s) == 0: return 0
    return conta_occorrenze_ric(c, s, 0) #cont delle occorrenze

def conta_occorrenze_ric(c, s_corr, occorrenze_corr):
    if len(s_corr) == 0: return occorrenze_corr
    if s_corr[0] == c:
        occorrenze_corr += 1
    return conta_occorrenze_ric(c, s_corr[1:], occorrenze_corr)
```
