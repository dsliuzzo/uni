## Esercizio 1
Si deve realizzare un metodo `public static boolean eRipetuto (AlberoB a, int x) {…}` che restituisce `true` se e solo se vi è almeno un nodo n nell’albero a tale che l’intero x appare sia nel sottoalbero sinistro che nel sottoalbero destro di n.

```python
from networkx.utils.random_sequence import random_weighted_sample

from alberi.alberibinari import AlberoBin

class es1(AlberoBin):
    def eRipetuto(self, a: AlberoBin, x: int) -> bool:
        if a is None:
            return False
        if es1.eRipetutoRic(a.sin,x) and es1.eRipetutoRic(a.des,x):
            return True
        return self.eRipetuto(a.sin,x) or self.eRipetuto(a.des,x)
    @staticmethod
    def eRipetutoRic(a,x):
        if a is None:
            return False
        return a.val==x or es1.eRipetutoRic(a.sin,x) or es1.eRipetutoRic(a.des,x)
```

### Complessità
#### Caso Peggiore
$$
T^W(n) \in \theta(n^2) \quad S^W(n) \in \theta(n)
$$
#### Caso Migliore

$$
T^B(n) \in \theta(1) \quad S^B(n) \in \theta(1)
$$

## Esercizio 2
verificare `almenoUnDoppio(self, a: AlberoBin, x: int)` che restituisce `True` se e solo se esiste `x` che appare sia in un nodo non foglia sia in un nodo foglia di `a`

```python
def almenoUnDoppio(a: AlberoBin, x: int) -> bool:
    if a is None:
        return False
    return xFoglia(a,x) and xNodo(a,x)

def xFoglia(a: AlberoBin, x: int) -> bool:
    if a is None:
        return False
    if a.sin is None and a.des is None and a.val == x:
        return True
    return xFoglia(a.sin,x) or xFoglia(a.des,x)

def xNodo(a: AlberoBin, x: int) -> bool:
    if a is None:
        return False
    if a.sin is None and a.des is None:
        return False
    return a.val==x or xNodo(a.sin,x) or xNodo(a.des,x)
```

### Complessità
#### Caso Peggiore
Nell'analisi del caso peggiore della complessità temporale è equivalente considerare un albero degenere o completo in quanto in entrambi i casi sia il metodo xFoglia che xNodo potrebbero dover scorrere tutti i nodi dell'albero ottenendo:

$$
T^W(n) \in (\theta(n) + \theta(n)) = \theta(n)
$$

Nell'analisi del caso peggiore della complessità temporale è necessario analizzare il caso di un albero degenere in quanto il metodo xFoglia alloca contemporaneamente n chiamate a se stessa.
$$
 S^W(n) \in \theta(n)
$$
#### Caso Migliore
Nel caso migliore si considera un albero il cui figlio sinistro della radice ha esattamente il valore x ed è un nodo foglia e il figlio destro della radice assume lo stesso valore x. Senza perdita di generalità, si assume che il figlio destro abbia come figli dei sottoalberi di dimensione generica. Per questo motivo, alla chiamata del metodo almeno un doppio si verifica in tempo costante che il metodo xFoglia restituisce subito true così come il metodo xNodo. Infatti, la complessità spaziale è costante proprio perchè le chiamate effettuate contemporaneamente sono 3, rendendo la complessità costante.
$$
T^B(n) \in \theta(1) \quad S^B(n) \in \theta(1)
$$

## Esercizio 3
Si definisca `foglieM(a: AlberoBin)` restituisce `True` se e solo se esiste solo una foglia dell'albero che ha valore $\geq 0$

```python
def foglieM(a:AlberoBin) -> bool:
    if a is None:
        return False
    return countFoglieM(a) == 1

def countFoglieM(a: AlberoBin) -> int:
    if a is None:
        return 0
    if a.sin is None and a.des is None:
        if a.val < 0:
            return 0
        return 1
    return countFoglieM(a.sin) + countFoglieM(a.des)
```

### Complessità
#### Caso Peggiore
Si considera un albero degenere che scorre tutti i nodi e alloca n chiamate contemporanee del metodo countFoglieM e non appena trova l'unica foglia può terminare.
$$
T^W(n) \in \theta(n) \quad S^W(n) \in \theta(n)
$$

#### Caso Migliore
Nel caso migliore, sia a livello temporale sia a livello spaziale, la complessità è costante nel caso in cui la radice abbia un figlio sinistro foglia e il figlio destro abbia un figlio sinistro foglia e un figlio destro i cui sottoalberi hanno dimensione generica. Se i corrispettivi valori delle foglie sono entrambe positive la condizione è violata  e restituisce subito False.
$$
T^B(n) \in \theta(n) \quad S^B(n) \in \theta(log(n))
$$

## Esercizio 4
Si restituisca l'altezza del nodo con valore `x` (se non è presente restituisce 0)

```python
def livelloX(a: AlberoBin, x: int) -> int:
    if a is None:
        return 0
    return livelloXRic(a,x,1)

def livelloXRic(a: AlberoBin, x: int, liv: int) -> int:
    if a is None:
        return 0
    if a.val == x:
        return liv
    return max(livelloXRic(a.sin,x,liv + 1), livelloXRic(a.des,x,liv + 1))

def altezzaX(a: AlberoBin, x: int) -> int:
    if a is None:
        return 0
    if a.val == x:
        return altezza(a)
    trovatoS = altezzaX(a.sin,x)
    if trovatoS != 0:
        return trovatoS
    return altezzaX(a.des,x)
def altezza(a: AlberoBin) -> int:
    if a is None:
        return 0
    return 1 + max(altezza(a.sin),altezza(a.des))
```
