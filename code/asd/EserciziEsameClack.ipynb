## Esercizio 1
Si deve realizzare un metodo `public static boolean eRipetuto (AlberoB a, int x) {…}` che restituisce `true` se e solo se vi è almeno un nodo n nell’albero a tale che l’intero x appare sia nel sottoalbero sinistro che nel sottoalbero destro di n.

```python
import sys

from networkx.utils.random_sequence import random_weighted_sample

from alberi.alberibinari import AlberoBin
from backtracking.backtracking import ProblemaBack
from grafi.algosugrafi import trovazero
from heap.heapmodificabile import HeapModificabile


class es1(AlberoBin):
    def eRipetuto(self, a: AlberoBin, x: int) -> bool:
        if a is None:
            return False
        if es1.eRipetutoRic(a.sin,x) and es1.eRipetutoRic(a.des,x):
            return True
        return self.eRipetuto(a.sin,x) or self.eRipetuto(a.des,x)
    @staticmethod
    def eRipetutoRic(a,x):
        if a is None:
            return False
        return a.val==x or es1.eRipetutoRic(a.sin,x) or es1.eRipetutoRic(a.des,x)
```

## Esercizio 2
verificare `almenoUnDoppio(self, a: AlberoBin, x: int)` che restituisce `True` se e solo se esiste `x` che appare sia in un nodo non foglia sia in un nodo foglia di `a`

```python
def almenoUnDoppio(a: AlberoBin, x: int) -> bool:
    if a is None:
        return False
    return xFoglia(a,x) and xNodo(a,x)

def xFoglia(a: AlberoBin, x: int) -> bool:
    if a is None:
        return False
    if a.sin is None and a.des is None and a.val == x:
        return True
    return xFoglia(a.sin,x) or xFoglia(a.des,x)

def xNodo(a: AlberoBin, x: int) -> bool:
    if a is None:
        return False
    if a.sin is None and a.des is None:
        return False
    return a.val==x or xNodo(a.sin,x) or xNodo(a.des,x)
```

### Complessità
#### Caso Peggiore
Nell'analisi del caso peggiore della complessità temporale è equivalente considerare un albero degenere o completo in quanto in entrambi i casi sia il metodo xFoglia che xNodo potrebbero dover scorrere tutti i nodi dell'albero ottenendo:

$$
T^W(n) \in (\theta(n) + \theta(n)) = \theta(n)
$$

Nell'analisi del caso peggiore della complessità temporale è necessario analizzare il caso di un albero degenere in quanto il metodo xFoglia alloca contemporaneamente n chiamate a se stessa.
$$
 S^W(n) \in \theta(n)
$$
#### Caso Migliore
Nel caso migliore si considera un albero il cui figlio sinistro della radice ha esattamente il valore x ed è un nodo foglia e il figlio destro della radice assume lo stesso valore x. Senza perdita di generalità, si assume che il figlio destro abbia come figli dei sottoalberi di dimensione generica. Per questo motivo, alla chiamata del metodo almeno un doppio si verifica in tempo costante che il metodo xFoglia restituisce subito true così come il metodo xNodo. Infatti, la complessità spaziale è costante proprio perchè le chiamate effettuate contemporaneamente sono 3, rendendo la complessità costante.
$$
T^B(n) \in \theta(1) \quad S^B(n) \in \theta(1)
$$

## Esercizio 5
Si definisca un metodo che restituisce true se esiste un nodo non foglia di valore v per cui la somma dei valori contenuti nel suo sottoalbero sinistro e destro sia minore dello stesso

```python
def somma(a: AlberoBin) -> bool:
    sommaTotale, trovato = sommaRic(a)
    return trovato

def sommaRic(a: AlberoBin) -> tuple:
    if a is None:
        return (0, False)

    sommaSin, trovatoSin = sommaRic(a.sin)
    sommaDes, trovatoDes = sommaRic(a.des)

    sommaTotale = a.val + sommaSin + sommaDes

    eFoglia = a.sin is None and a.des is None
    trovatoQui = (not eFoglia) and (a.val < sommaSin + sommaDes)

    trovato = trovatoSin or trovatoDes or trovatoQui

    return (sommaTotale, trovato)
```

# Scheduling delle attività
Caso pratico di utilizzo di un grafo aciclico orientato (DAG).

- Ogni nodo del grafo rappresenta un'attività da svolgere.
- Ogni arco rappresenta una relazione di propedeuticità tra due attività (es. A→B significa "A deve terminare prima che B possa iniziare").
- Ogni attività ha un proprio tempo di esecuzione.
- Hai a disposizione un numero illimitato di esecutori (quindi più attività possono partire in parallelo, appena le loro propedeuticità sono soddisfatte).
- Si parte al tempo 0.

Domanda: qual è il tempo minimo necessario per completare tutte le attività, rispettando i vincoli di propedeuticità?

```python
from grafi.grafi import Grafo

class Scheduler(Grafo):
    def trovaZero(self, gradi: list[int], rimossi: list[bool]) -> list[tuple[int,int]]:
        res: list[tuple[int,int]] = []
        for i in range (len(gradi)):
            if gradi[i] == 0 and not rimossi[i]:
                rimossi[i] = True
                res.append((i,i))
        return res

    def min_scheduler(self, g: Grafo) -> int:
        tempo: int = 0
        exe: list[tuple[int,int]] = []
        gradi: list[int] = [0 for i in range(g.n)]
        rimossi: list[bool] = [False for i in range(g.n)] # nodo i-esimo è stato rimosso se True
        for x,y in g.archi():
            gradi[y] += 1
        curr = self.trovaZero(gradi,rimossi)
        exe.extend(curr)
        while curr != (-1 , -1):
            curr = min(exe, key=lambda x: x[1])
            tempo += curr[1]
            for i in range(len(exe)):
                exe[i] = (exe[i][0], exe[i][1] - curr[1])
            for ad in g.adiacenti(curr):
                gradi[ad] -= 1
            curr = self.trovaZero(gradi,rimossi)
            exe.extend(curr)
        return tempo
```

# Esercizio
Dato un grafo non orientato pesato (pesi = costo di percorrenza), calcolare la centralità di ogni nodo, definita come il numero di cammini minimi (tra tutte le coppie di nodi) che passano per quel nodo come nodo intermedio.

L'idea è identificare i nodi "cruciali" per la connettività del grafo: rimuovendo un nodo ad alta centralità, il grafo diventa più difficile da attraversare (applicazioni: interdizione militare di uno snodo, chiusura di un nodo per rallentare un contagio).

Richiesta: modificare l'algoritmo di Floyd-Warshall in modo da, oltre alle distanze minime, tenere traccia anche del numero di cammini minimi che passano per ciascun nodo intermedio, tenendo conto che possono essercene più di uno (potenzialmente esponenziali) per coppia.

```python
def FloydVar (g: Grafo) -> tuple[list[list[int]],dict]:

    distanze: list[list[int]] = [[sys.maxsize for _ in range(g.n)] for _ in range(g.n)]
    for a in range(g.n):
        distanze[a][a] = 0
    for x,y,peso in g.archi():
        distanze[x][y] = peso
    for k in range(g.n):
        for x in range(g.n):
            for y in range(g.n):
                if distanze[x][y] > distanze[x][k] + distanze[k][y]:
                    distanze[x][y] = distanze[x][k] + distanze[k][y]
    central: dict = {}
    for k in range(g.n):
        for x in range(g.n):
            for y in range(g.n):
                if distanze[x][y] == distanze[x][k] + distanze[k][y] and k != x and k != y:
                    central[k] += 1
    return distanze, central

```

## Distanza minima tra due stringhe
Calcolo della distanza minima tra due stringhe potendo effettuare l'operazione di swap.
Lo swap ammesso è solo quello tra **caratteri adiacenti** (es. "ab" → "ba"), non uno scambio libero tra posizioni arbitrarie.
**Perché il vincolo è necessario:**
- Concatenare swap sulle stesse posizioni (o posizioni vicine) equivarrebbe a far "scivolare" un carattere lungo la stringa a costo 1 per passo.
- Questo simula di fatto un riordinamento generale della stringa, che è il caso esponenziale da evitare.
- Vietando gli swap a catena, ogni swap resta un'operazione **isolata e locale**.
- Si preserva così la sottostruttura ottima: ogni sottoproblema dipende da un numero fisso di sottoproblemi precedenti, senza dover tenere traccia della storia delle operazioni — condizione necessaria per una DP in tempo polinomiale.

```python
def editdistance(s1: String, s2: String) -> int:
    distanze: list[list[int]] = [[0 for _ in range(len(s2) + 1)] for _ in range(len(s1) + 1)]
    for i in range(len(s1) + 1):
        for j in range(len(s2) + 1):
            if i == 0:
                distanze[i][j] = j
            elif j == 0:
                distanze[i][j] = i
            else:
                distanze[i][j] = min(distanze[i-1][j] + 1, distanze[i][j-1] + 1, (0 if s1[i]==s2[j] else 1) + distanze[i-1][j-1])
    return distanze[len(s1) - 1][len(s2) - 1]
```

## Esercizio con Backtracking

Dato un grafo $G = (V, E)$, un **dominating set** (insieme dominante) è un sottoinsieme $S \subseteq V$ tale che:

$$\forall v \in V,\quad v \in S \ \lor \ \exists u \in S : (u,v) \in E$$

Cioè, per **ogni** nodo $v$ del grafo (non solo quelli in $S$), deve valere almeno una delle due condizioni:

1. **$v \in S$** — il nodo appartiene direttamente all'insieme, oppure
2. **$\exists u \in S$ t.c. $(u,v) \in E$** — esiste un nodo $u \in S$ collegato a $v$ tramite un arco (un "vicino" in $S$)
