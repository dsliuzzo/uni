# Cambiamo argomento
Abbiamo un grafo G non orientato e pesato.
G è un grafo sociale dove i nodi sono i utenti e gli archi sono relazioni di amicizia.
Il peso dell'arco è la forza dell'amicizia✨.
Bisogna implementare `forza(G,k,h)` restituisce gli id degli utenti che hanno almeno `k` amici collegati da archi di peso $\geq$ `h`.

```python
def forza(G, k : int,h : int):
    res = []
    for i in range(G.n):
        count = 0
        for a in G.adiacenti(i): # restituisce una lista di elementi composti da nodi e peso
            if a.peso >= h:
                count += 1
                if count >= k:
                    res.append(i)
                    break
    return res
```

## Complessità
### Matrice di adiacenza
**Caso peggiore**
Doppio for innestato, in entrambi si arriva ad $n$, quindi
$$CTP(n,m) = \Theta(n^2)$$
la spaziale invece si considera solo la creazione della lista, nel peggiore dei casi la lista è grande quanto $n$
$$CSP(n,m) = \Theta(n)$$
**Caso migliore**
$$CSM(n,m) = \Theta(1)$$
$$CTM(n,m) = \Theta(n)$$
### Lista di adiacenza
la lista restituita da `G.adiacenti(i)` restituisce una lista
**Caso migliore**
$$CTP(n,m) = \Theta(n^2)$$
per la spaziale non cambia tra lista e matrice, conta solo res, quindi la complessità è uguale
$$CSP(n,m) = \Theta(n)$$
**Caso peggiore**
$$CSM(n,m) = \Theta(1)$$
$$CTM(n,m) = \Theta(n)$$

# Esercizio 2
Grafo non orientato e non pesato in cui i nodi sono gli utenti e gli archi sono le relazioni di amicizia.
Definire `popular(G)` che restituisce l'utente che ha il maggior numero di amici e amici di amici

```python
# mia soluzione sbagliata
def popular(G):
    res = None
    max_amici = 0
    for i in range(G.n):
        nAmici : int = contaAmici(i,G)
        for a in G.adiacenti(i):
            nAmici += contaAmici(a, G)
        if nAmici > max_amici:
            res = i
            max_amici = nAmici
    return res

def contaAmici(nodo, G):
    return len(G.adiacenti(nodo))
```

```python
def popular_nodo(G,i):
    visitato = [False for i in range(G.n)]
    conta = 0
    visitato[i] = True
    for j in G.adiacenti(i):
        if not visitato[j]:
            visitato[j] = True
            conta += 1
        for h in G.adiacenti(j):
            if not visitato[h]:
                if not visitato[h]:
                    visitato[h] = True
                    conta += 1
    return conta

def popular(G):
    nodo_max = None
    valore_max = -1
    for i in range(G.n):
        valore = popular_nodo(G,i)
        if valore > valore_max:
            valore_max = valore
            nodo_max = i
    return nodo_max
```

## Complessità
### Lista di adiacenza - Matrice di adiacenza
**Caso migliore**
$$CTM(n,m) = \Theta(n^2)$$
$$CSM(n,m) = \Theta(n)$$
**Caso peggiore**
$$CTP(n,m) = \Theta(n^3)$$
$$CSP(n,m) = \Theta(n)$$
