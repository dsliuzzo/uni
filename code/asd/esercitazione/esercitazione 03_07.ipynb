# Colorare un grafo
Grafo non orientato, assegnare un colore a ogni nodo, facendo in modo che non ci siano colori tra nodi adiacenti.

Scegliere tra un numero `c` di colori, b,r,v `colori=[b = 0, r = 1, v = 2]`

`liv =` nodo da colorare

`scelta =` scelta del colore dalla lista `colori`

`vincolo =` colore del nodo sul livello deve essere diverso dagli adiacenti

`soluzione[i] =` colore assegnato al nodo

```python
from alberiBinari import AlberoBin


class coloraGrafo(ProblemaBack):
    def __init__(self, g:grafoNO, num_color:int):
        self.g:grafoNO = g
        self.num_color = num_color
        self.colore:list[int] = [-1 for _ in range(g.n)]
        self.soluzione:list[int] = []

    def primaScelta(self, liv:int) -> bool:
        # assegnare il primo colore, quindi 0
        if liv == self.g.n: return False # abbiamo esaurito le scelte
        self.colore[liv] = 0
        return True

    def successivaScelta(self, liv:int) -> bool:
        # se il colore che ho assegnato non è l'ultimo ne posso assegnare un altro
        if self.colore[liv] < self.num_color -1:
            self.colore[liv] += 1
            return True
        return False # per il nodo liv abbiamo provato tutti i colori, quindi non è una soluzione corretta

    def verificaVincoli(self, liv:int) -> bool:
        # devo verificare se esiste un arco tra il nodo liv e i nodi già colorati (<liv) e i due nodi hanno lo stesso colore
        for i in range(liv):
            if self.g.arco(i,liv) and self.colore[i] == self.colore[liv]:
                return False
        return True

    def soluzioneCompleta(self, liv:int) -> bool:
        # la soluzione è completa se ho verificato tutti i vicoli e sono arrivato all'ultimo nodo
        return liv == self.g.n-1 # basta verificare quindi che liv è l'ultimo nodo (se non sono verificati i vincoli viene bloccato prima)

    def costruisciSoluzione(self, liv:int):
        for x in self.colore:
            self.soluzione.append[x]
```

# N regine
Bisogna posizionare `n` regine su una scacchiera `n*n` in modo che non si mangino tra loro
`liv =` riga
`output =` matrice (per semplicità lo vediamo come una lista in cui gli indici sono le righe e ogni cella contiene la posizione della colonna e poi in costruisci soluzione possiamo creare la matrice
`vincolo =` non si devono essere sulle stesse righe (abbiamo risolto con l'uso della lista), colonne e diagonali

```python
class nRegine(ProblemaBack):
    def __init__(self, num_regine:int):
        self.num_regine = num_regine
        self.colonna:list[int] = [-1 for _ in range(num_regine)]
        self.soluzione:list[bool][bool] = [[False for _ in range(num_regine)] for _ in range(num_regine)] # va inizializzata per ogni riga

    def primaScelta(self, liv:int) -> bool:
        # assegnare la prima scelta, quindi 0
        if liv == self.num_regine: return False # abbiamo esaurito le scelte
        self.colonna[liv] = 0
        return True

    def successivaScelta(self, liv:int) -> bool:
        # se la colonna che ho assegnato non è l'ultima ne posso assegnare un'altra
        if self.colonna[liv] < self.num_regine -1:
            self.colonna[liv] += 1
            return True
        return False # abbiamo provato tutte le possibili colonne

    def verificaVincoli(self, liv:int) -> bool:
        # devo verificare se c'è un'altra regina che sta sulla stessa colonna
        for i in range(liv):
            # capisco se sono nella diagonale perché la distanza tra le righe e la distanza tra le colonne è uguale
            if self.colonna[i] == self.colonna[liv] or abs(self.colonna[i] - self.colonna[liv]) == liv-i:
                return False
        return True

    def soluzioneCompleta(self, liv:int) -> bool:
        # la soluzione è completa se ho verificato tutti i vicoli e sono arrivato all'ultima riga
        return liv == self.num_regine-1

    def costruisciSoluzione(self, liv:int):
        for i in range(self.num_regine):
            self.soluzione[i][self.colonna[i]] = True
```

# Prova di esame - verifica somma
## Primo es
Restituire `True` se la somma del contenuto delle foglie di `a` è uguale alla somma dei nodi a livello `l` di `a`


```python
def verificaSomma(a: AlberoBin, l: int) -> bool:
    if a is None: return True
    somma = [0,0]
    verifica(a,l,0,somma)
    return somma[0] == somma[1]

def verifica(a, l: int, curr: int, somma):
    if a is None: return
    if curr == l: somma[0] += a.val
    if a.sin is None and a.des is None: somma[1] += a.val
    verifica(a.des,l,curr+1,somma)
    verifica(a.sin,l,curr+1,somma)
```
