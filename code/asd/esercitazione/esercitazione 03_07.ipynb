# Colorare un grafo
Assegnare un colore a ogni nodo, facendo in modo che non ci siano colori tra nodi adiacenti.

Scegliere tra b,r,v `colori=[b = 0, r = 1, v = 2]`

`liv =` nodo da colorare

`scelta =` scelta del colore dalla lista `colori`

`vincolo =` colore del nodo sul livello deve essere diverso dagli adiacenti

`soluzione =` lista di colori per ogni nodo (indice è "nome" del nodo)
